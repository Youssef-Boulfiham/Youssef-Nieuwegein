```python
def set_positions_friends():
    activities = ["school", "vrienden thuis", "vrije tijd"]
    
    # Example coordinates for each activity (replace with actual values)
    coordinates = [
        [384, 336, 350, 224],  # (x, y, x1, y1) for "school"
        [270, 462, 240, 380],   # (x, y, x1, y1) for "vrienden thuis"
        [480, 624, 432, 524]   # (x, y, x1, y1) for "vrije tijd"
    ]

    all_positions = {}

    for i, activity in enumerate(activities):
        x, y, x1, y1 = coordinates[i]
        n = 12  # Number of divisions

        # Create y values based on division count
        y_values = [y + round(j * (y1 - y) / n) for j in range(n + 1)]
        
        # Generate positions for the two vertical lines
        positions = {
            "line1": [(x, yi) for yi in y_values],
            "line2": [(x1, yi) for yi in y_values]
        }
        
        all_positions[activity] = positions
        print(f"{activity} positions:", positions)

    return all_positions

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
# oad the image
image_path = "/Users/youssefboulfiham/PycharmProjects/pythonProject/Youssef-Nieuwegein/graphics/enviroment_activity.png"
image = Image.open(image_path).convert("RGB")

# Get image dimensions
width, height = image.size

# Get all positions
all_positions = set_positions_friends()

# Create plot
plt.figure(figsize=(10, 8))
plt.imshow(image)  # Display the image as background

# Define colors for each activity
colors = {"school": "red", "vrienden thuis": "blue", "vrije tijd": "green"}

# Plot vertical lines for each activity
for activity, positions in all_positions.items():
    color = colors.get(activity, "black")  # Default to black if not defined

    # Extract and plot the two vertical lines
    for line in ["line1", "line2"]:
        x_vals, y_vals = zip(*positions[line])
        plt.plot(x_vals, y_vals, marker="o", linestyle="-", color=color, label=activity if line == "line1" else None)

# Labels and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.title("Activity Positions on Image")

# Display the plot
plt.show()
```
